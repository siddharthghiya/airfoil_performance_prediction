```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_loss=np.load("train loss.npy")
test_loss= np.load("test loss.npy")
```

```python
epochs= np.linspace(1,100,100)
plt.figure(figsize=(8,5))
plt.title("Train and Test Loss vs Epochs",fontsize='x-large')
plt.plot(epochs,train_loss,label='Train')
plt.plot(epochs,test_loss,label='Test')
plt.xlabel("Epochs",fontsize='x-large')
plt.ylabel("Loss",fontsize='x-large')
plt.legend(fontsize='x-large')
plt.savefig('./Results/CNN_loss_plot.png',dpi=288)
```

```python
final_data=np.load('Foil_CNN_prediction_99.npy')
final_data = final_data.reshape(-2,6)
accuracy = 1 - np.abs(final_data[:,3].astype(float) - final_data[:,5].astype(float))/np.maximum(np.abs(final_data[:,3].astype(float)),np.abs(final_data[:,5].astype(float)))
net_accuracy = np.average(accuracy)
print('net accuracy is',net_accuracy*100)
```

```python
epochs= np.linspace(1,5,5)
tanh_train_loss=np.log(np.load('train loss tanh.npy'))
relu_train_loss=np.log(np.load('train loss relu.npy'))
plt.figure(figsize=(8,5))
plt.title("Tanh vs ReLU Train Loss",fontsize='x-large')
plt.plot(epochs,tanh_train_loss,label='tanh')
plt.plot(epochs,relu_train_loss,label='Relu')
plt.xlabel("Epochs",fontsize='x-large')
plt.ylabel("Log Loss",fontsize='x-large')
plt.legend(fontsize='x-large')
plt.savefig('./Results/tanvsRelutrain_loss_plot.png',dpi=288)
```

```python
epochs= np.linspace(1,5,5)
tanh_test_loss=np.log(np.load('test loss tanh.npy'))
relu_test_loss=np.log(np.load('test loss relu.npy'))
plt.figure(figsize=(8,5))
plt.title("Tanh vs ReLU Test Loss",fontsize='x-large')
plt.plot(epochs,tanh_test_loss,label='tanh')
plt.plot(epochs,relu_test_loss,label='Relu')
plt.xlabel("Epochs",fontsize='x-large')
plt.ylabel("Log Loss",fontsize='x-large')
plt.legend(fontsize='x-large')
plt.savefig('./Results/tanvsRelutest_loss_plot.png',dpi=288)
```
